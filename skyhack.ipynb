```python
def calculate_satisfaction_ratio_by_origin(merged_df2, N=20):
    # Calculate the top N origin stations by flight count
    top_n_origin_stations = merged_df2['origin_station_code'].value_counts().nlargest(N)

    # Create a subset dataframe with top N origin stations
    df_top_n = merged_df2[merged_df2['origin_station_code'].isin(top_n_origin_stations.index)]

    # Calculate the satisfaction ratio for each origin station
    satisfaction_ratio = df_top_n.groupby('origin_station_code')['satisfaction_type'].value_counts(normalize=True).unstack()
    if 'satisfied' in satisfaction_ratio.columns and 'dissatisfied' in satisfaction_ratio.columns:
        satisfaction_ratio['Satisfaction Ratio (Satisfied / Dissatisfied)'] = satisfaction_ratio['satisfied'] / satisfaction_ratio['dissatisfied']

    # Reset index for better formatting
    satisfaction_ratio.reset_index(inplace=True)

    return satisfaction_ratio

# Call the function with your merged_df2 dataframe
satisfaction_ratio_table = calculate_satisfaction_ratio_by_origin(merged_df2)
print(satisfaction_ratio_table.sort_values(by='Dissatisfied'))
```

```python
import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# Load your dataframe here
# df = ...

# Text Preprocessing
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    words = [stemmer.stem(word) for word in words if word not in stop_words]
    return words

df['tokenized_text'] = df['verbatim_text'].apply(preprocess_text)

# Word Frequency Analysis
all_words = [word for words in df['tokenized_text'] for word in words]
word_freq = Counter(all_words)

# Visualize Most Common Words
num_common_words = 20
common_words = word_freq.most_common(num_common_words)
words, counts = zip(*common_words)

plt.figure(figsize=(10, 6))
plt.barh(words, counts)
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title(f'Most Common {num_common_words} Words in Verbatim Text')
plt.show()

# Identify Pain Points
pain_point_keywords = ['delay', 'cancel', 'uncomfortable', 'issue', 'problem', 'bad', 'poor', 'disappoint']
pain_point_counts = {word: word_freq[word] for word in pain_point_keywords}

print("Pain Point Keywords and Their Frequencies:")
for word, count in pain_point_counts.items():
    print(f"{word}: {count}")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt


# Step 1: Filter Relevant Data
food_beverage_dissatisfaction = df[(df['satisfaction_type'] == 'Dissatisfied') &
                                   (df['verbatim_text'].str.contains('food|beverage', case=False))]

# Step 2: Tokenize and Process Words
stop_words = set(stopwords.words('english'))
word_freq = Counter()

for feedback in food_beverage_dissatisfaction['verbatim_text']:
    words = word_tokenize(feedback.lower())  # Tokenize and convert to lowercase
    words = [word for word in words if word.isalnum() and word not in stop_words]  # Remove stopwords and punctuation
    word_freq.update(words)

# Step 3: Identify Food and Beverage Related Words
food_beverage_keywords = ['salad', 'dressing', 'meal', 'dish', 'flavor', 'taste', 'portion', 'presentation', 'beverage', 'drink']

dissatisfaction_keywords = {}

for keyword in food_beverage_keywords:
    if keyword in word_freq:
        dissatisfaction_keywords[keyword] = word_freq[keyword]

# Step 4: Visualize Keywords and Their Frequencies
plt.bar(dissatisfaction_keywords.keys(), dissatisfaction_keywords.values())
plt.xlabel('Food and Beverage Keywords')
plt.ylabel('Frequency')
plt.title('Frequency of Food and Beverage Keywords in Dissatisfaction Feedback')
plt.xticks(rotation=45)
plt.show()
```
